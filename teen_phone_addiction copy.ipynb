```python
df.drop(columns=['ID', "Name","Location"], inplace=True)
df
```

```python
# Check for missing values 
missing = df.isnull().sum()
missing = missing[missing > 0]

if not missing.empty:
    print("Missing values found:\n", missing)
    for col in df.columns:
        if df[col].dtype == 'object':
            df[col] = df[col].fillna(df[col].mode()[0])
        else:
            df[col] = df[col].fillna(df[col].mean())
    print("\nMissing values handled.")
else:
    print(" No missing values detected.")
```

```python
bins=[0, 5, 8, 10]
labels=['Low', 'Moderate', 'High']
df['Addiction_Level_New'] = pd.cut(df['Addiction_Level'], bins=bins, labels=labels)

df.drop(columns=['Addiction_Level'], inplace=True)

print(df['Addiction_Level_New'].value_counts())
```

```python
df["Addiction_Level_New"].value_counts(normalize=True).plot(kind='bar', color='skyblue', title='Addiction Level Distribution')
```

```python
# split the dataset into features and target variable

X = df.drop(columns=['Addiction_Level_New'])
y = df["Addiction_Level_New"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=8, stratify=y) 
```

```python
# preprocess data
from sklearn.compose import ColumnTransformer
from sklearn.discriminant_analysis import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder


numeric_features = X_train.select_dtypes(include=['int64', 'float64']).columns
categorical_features = X_train.select_dtypes(include=['object']).columns.to_list()

numeric_transformer = Pipeline(steps=[
    ('scaler', StandardScaler())
])
categorical_transformer = Pipeline(steps=[
    ('onehot', OneHotEncoder(handle_unknown='ignore', sparse=False))
])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ]
)
X_train_encoded = preprocessor.fit_transform(X_train)
X_test_encoded = preprocessor.transform(X_test)
```

```python
# use SMOTE for balancing Addiction Level New Dataset

from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=8)
X_train_balanced, y_train_balanced = smote.fit_resample(X_train_encoded, y_train)

print(y_train_balanced.value_counts())
```

```python
print("NaNs in X_train_processed:", np.isnan(X_train_balanced).sum())
print("NaNs in X_test_processed:", np.isnan(X_test_encoded).sum())
print("NaNs in y_train:", pd.Series(y_train_balanced).isna().sum())
print("NaNs in y_test:", pd.Series(y_test).isna().sum())
```

I then dropped the old (Addiction_Level) and use (Addiction_Level_New) as my target. After that, i split it into features and target variables. I used stratify=y to ensure that the class distribution was consistent in both training and testing sets, especially because i have class imbalance!

Last but not least, i did one last check for my train and test for NaN values.

<h1><font color=red> (b) Model building and evaluation of model </font></h1>

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier


from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import GridSearchCV
```

```python
# Define the models to be used

models = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "Decision Tree": DecisionTreeClassifier(random_state=8),
    "Random Forest": RandomForestClassifier(n_estimators = 100, random_state=8),
    "Gradient Boosting": GradientBoostingClassifier(random_state=8)
}

# Train and evaluate each model
results = []
for name, model in models.items():
    model.fit(X_train_balanced, y_train_balanced)
    y_pred = model.predict(X_test_encoded)
    acc = accuracy_score(y_test, y_pred)
    report = classification_report(y_test, y_pred, output_dict=True)
    results.append({
        "Model": name,
        "Accuracy": acc,
        "Precision (macro avg)": report['macro avg']['precision'],
        "Recall (macro avg)": report['macro avg']['recall'],
        "F1-Score (macro)": report['macro avg']['f1-score']
    })
    
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=model.classes_, yticklabels=model.classes_)
    plt.title(f'Confusion Matrix for {name}')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.tight_layout()
    plt.show()

results_df = pd.DataFrame(results)
display(results_df)
    
```

```python
# hyperparameter tuning for logistic regression using GridSearchCV

param_grid_lr = [
    {
        'penalty': ['l2'],
        'C': [0.001, 0.01, 0.1, 1, 10, 100],
        'solver': ['lbfgs', 'newton-cg', 'sag']
    },
    {
        'penalty': ['l1'],
        'C': [0.001, 0.01, 0.1, 1, 10, 100],
        'solver': ['liblinear', 'saga']
    },
    {
        'penalty': ['elasticnet'],
        'C': [0.001, 0.01, 0.1, 1, 10, 100],
        'solver': ['saga'],
        'l1_ratio': [0.1, 0.5, 0.9] 
    },
]

grid_search_lr = GridSearchCV(
    LogisticRegression(max_iter=100), 
    param_grid_lr, 
    cv=5, 
    scoring='f1_macro', 
    n_jobs=-1
)

grid_search_lr.fit(X_train_balanced, y_train_balanced)
best_params_lr = grid_search_lr.best_params_
print("Best parameters for Logistic Regression:", best_params_lr)

best_lr=grid_search_lr.best_estimator_
```

```python
best_lr.fit(X_train_balanced, y_train_balanced)
y_pred = best_lr.predict(X_test_encoded)

# Evaluation
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
# hyperparameter tuning for Decision tree using GridSearchCV
param_grid_dt = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [5, 10, 20, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
}
grid_search_dt = GridSearchCV(
    estimator=DecisionTreeClassifier(random_state=8),
    param_grid=param_grid_dt,
    scoring='f1_macro',
    cv=5,
    verbose=1,
    n_jobs=-1
)

grid_search_dt.fit(X_train_balanced, y_train_balanced)
best_params_dt = grid_search_dt.best_params_
print("Best parameters for Decision Tree: ", best_params_dt)

best_dt = grid_search_dt.best_estimator_
```

```python
best_dt.fit(X_train_balanced, y_train_balanced)
y_pred = best_dt.predict(X_test_encoded)

# Evaluation
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
# hyperparameter tuning for Random Forest using GridSearchCV
param_grid_rf = {
    'n_estimators': [100, 300, 500],
    'max_features': ['sqrt', 'log2'],
    'max_depth': [10, 20, 30, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
}

# initialise the GridSearchCV object for RandomForestClassifier
grid_search_rf = GridSearchCV(
    estimator=RandomForestClassifier(
        random_state=8, 
        class_weight='balanced'
    ),
    param_grid=param_grid_rf,
    cv=5,
    scoring='f1_macro',
    n_jobs=-1, 
    verbose=1
)

# fit the grid search to the data
grid_search_rf.fit(X_train_balanced, y_train_balanced)

# get the best parameters
best_params_rf = grid_search_rf.best_params_
print('Best parameters for RandomForest:', best_params_rf)

# get the best estimator
best_rfc = grid_search_rf.best_estimator_
```

```python
best_rfc.fit(X_train_balanced, y_train_balanced)
y_pred = best_rfc.predict(X_test_encoded)

# Evaluation
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
# hyperparameter tuning for Gradient Boosting using GridSearchCV
param_grid_gb = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 5, 7],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
}
grid_search_gb = GridSearchCV(
    estimator=GradientBoostingClassifier(random_state=8), param_grid=param_grid_gb, 
    cv=5, 
    scoring='f1_macro', 
    n_jobs=-1, 
    verbose=1
)
grid_search_gb.fit(X_train_balanced, y_train_balanced)
best_params_gb = grid_search_gb.best_params_
print('Best parameters for Gradient Boosting:', best_params_gb)
best_gbc = grid_search_gb.best_estimator_
```

```python
best_gbc.fit(X_train_balanced, y_train_balanced)
y_pred = best_gbc.predict(X_test_encoded)

# Evaluation
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
# Define the models to be used
models = {
    "Logistic Regression": grid_search_lr.best_estimator_,
    "Decision Tree": grid_search_dt.best_estimator_,
    "Random Forest": grid_search_rf.best_estimator_,
    "Gradient Boosting": grid_search_gb.best_estimator_
}

# Train and evaluate each model (all model at once)
results = []
for name, model in models.items():
    model.fit(X_train_balanced, y_train_balanced)
    y_pred = model.predict(X_test_encoded)
    acc = accuracy_score(y_test, y_pred)
    report = classification_report(y_test, y_pred, output_dict=True)
    results.append({
        "Model": name,
        "Accuracy": acc,
        "Precision (macro avg)": report['macro avg']['precision'],
        "Recall (macro avg)": report['macro avg']['recall'],
        "F1-Score (macro)": report['macro avg']['f1-score']
    })
    
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=model.classes_, yticklabels=model.classes_)
    plt.title(f'Confusion Matrix for {name}')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.tight_layout()
    plt.show()

results_df = pd.DataFrame(results)
display(results_df)
```

```python
# check feature importance
feat_names = preprocessor.get_feature_names_out()
importances = best_lr.coef_[0]
imp_df = pd.DataFrame({'Feature': feat_names, 'Importance': importances})
print(imp_df.sort_values(by='Importance', ascending=False).head(15))
```

```python
# combine into stacked model
from sklearn.ensemble import VotingClassifier
models = {
    "Logistic Regression": best_lr,
    "Decision Tree": best_dt,
    "Random Forest": best_rfc,
    "Gradient Boosting": best_gbc
}
voting_clf = VotingClassifier(estimators=list(models.items()), voting='soft')
voting_clf.fit(X_train_balanced, y_train_balanced)
y_pred = voting_clf.predict(X_test_encoded)
# Evaluation
print("Accuracy of Stacked Model:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title("Confusion Matrix for Stacked Model")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
# Use the best logistic regression model and balanced, encoded data
from sklearn.model_selection import learning_curve

train_sizes, train_scores, test_scores = learning_curve(
    best_lr, X_train_balanced, y_train_balanced,
    cv=5, scoring='accuracy',
    train_sizes=np.linspace(0.1, 1.0, 10),
    n_jobs=-1
)

train_mean = train_scores.mean(axis=1)
test_mean = test_scores.mean(axis=1)

plt.figure(figsize=(10, 6))
plt.plot(train_sizes, train_mean, 'o-', label='Training Accuracy')
plt.plot(train_sizes, test_mean, 'o-', label='Validation Accuracy')
plt.xlabel('Training Set Size')
plt.ylabel('Accuracy')
plt.title('Learning Curve - Logistic Regression on Balanced Dataset')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

<h1><font color="red">Justification for (b) Model Training and Evaluation:</font></h1>

Firstly, i trained 4 models, Logistic Regression, Decision Tree, Random Forest and Gradient Boosting. Each of them are evaluated using metrics like Accuracy, Precision, Recall and f1-score. After that i also plotted confusion matrics for visual inspection of misclassification. 

Secondly, i did hyperparameter tuning for all 4 models. I implemented GridSearchCV for all 4 models to find the one with the best performing hyperparameters. I used scoring="f1-micro" for all models to ensure fair evaluation across all classes including the minority ones

Thirdly, i trained all 4 models once again after tuning to make sure everything was correctly done and compare the model using a summary table. The final model selection was Logistic Regression as the overall of that model was the best out of all 4. 

Aside from that, i also check for feature importance of the logistic regression to understand which variable influences the model the most. 

In hope to further improve my classification performance, I tried combining all 4 models into a stacked model using soft voting ensemble. The evaluation metrics and the confusion metrics was crafted but the overall result of the stacked model was not as good as my Logistic Regression model alone hence i decided to just use my logistic regression model. 

Last but not least, i then did a learning curve of my logistic regression to visualise the relationship between the size of the training data and model performance. The curve displays both the training accuracy and validation accuracy across increasing training set sizes. The performance pattern suggest that the logistic regression model is well and not overfitting. It also confirms my stand that logistic regression is best suited for my dataset.
